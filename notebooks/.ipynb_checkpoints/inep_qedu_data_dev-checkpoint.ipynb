```python
## extract_explore.py — extração exploratória do Censo Escolar para Uauá-BA
## Objetivo: baixar o Censo, filtrar o município e inspecionar como os dados vêm

import requests
import zipfile
import io
import ssl
import pandas as pd
from requests.adapters import HTTPAdapter
from urllib3.util.ssl_ import create_urllib3_context

## --- Constantes do projeto ---
URL_CENSO_2025 = "https://download.inep.gov.br/dados_abertos/microdados_censo_escolar_2025.zip"
CO_MUNICIPIO_UAUA = 2932408  ## código IBGE de Uauá-BA
CSV_INTERNO = "microdados_censo_escolar_2025/dados/microdados_ed_basica_2025.csv"

## Cabeçalho de navegador — o servidor do INEP bloqueia requisições "cruas"
HEADERS = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64)"}


## --- Adapter SSL para o servidor legado do INEP ---
## O servidor do INEP usa uma configuração TLS antiga que o OpenSSL novo
## (SECLEVEL=2 por padrão) recusa, causando UNEXPECTED_EOF_WHILE_READING.
## Rebaixar para SECLEVEL=1 destrava o handshake.
class INEPAdapter(HTTPAdapter):
    def init_poolmanager(self, *args, **kwargs):
        ctx = create_urllib3_context()
        ctx.check_hostname = False
        ctx.verify_mode = ssl.CERT_NONE
        ctx.set_ciphers("DEFAULT@SECLEVEL=1")
        kwargs["ssl_context"] = ctx
        return super().init_poolmanager(*args, **kwargs)


def criar_sessao_inep():
    """Cria uma sessão HTTP compatível com o TLS legado do INEP."""
    sessao = requests.Session()
    sessao.mount("https://", INEPAdapter())
    return sessao


## --- 1. Download do ZIP ---
print("Baixando o Censo 2025 (arquivo grande, pode levar alguns minutos)...")
sessao = criar_sessao_inep()
resp = sessao.get(URL_CENSO_2025, headers=HEADERS, timeout=600)
resp.raise_for_status()
print(f"Download concluído: {len(resp.content)/1024/1024:.1f} MB")

## --- 2. Abrir o ZIP em memória e ler só o CSV de escolas ---
with zipfile.ZipFile(io.BytesIO(resp.content)) as z:
    ## Confirma o nome exato do arquivo interno
    print("\nArquivos no ZIP:")
    for nome in z.namelist():
        if nome.endswith(".csv"):
            print("  ", nome)

    with z.open(CSV_INTERNO) as f:
        ## Lê filtrando: só carrega as colunas úteis e só as linhas de Uauá
        df = pd.read_csv(
            f,
            sep=";",
            encoding="latin-1",       ## pegadinha clássica do INEP
            decimal=",",
            low_memory=False,
        )

## --- 3. Filtrar Uauá ---
uaua = df[df["CO_MUNICIPIO"] == CO_MUNICIPIO_UAUA].copy()

print(f"\n{'='*60}")
print(f"Total de escolas no Brasil: {len(df):,}")
print(f"Total de escolas em Uauá-BA: {len(uaua)}")
print(f"{'='*60}\n")

## --- 4. Inspecionar a estrutura ---
print("Colunas disponíveis (primeiras 30):")
print(list(df.columns[:30]))

print("\nEscolas de Uauá encontradas:")
colunas_vista = ["CO_ENTIDADE", "NO_ENTIDADE", "TP_DEPENDENCIA", "TP_LOCALIZACAO"]
print(uaua[colunas_vista].to_string(index=False))

## Procurar pelo nome oficial atual — termos mais distintivos
print("\nProcurando Colégio Estadual do Campo Integração Lagoa do Pires:")
mask = uaua["NO_ENTIDADE"].str.contains("INTEGRA|LAGOA DO PIRES|LAGOA", case=False, na=False)
print(uaua[mask][colunas_vista].to_string(index=False))
```
